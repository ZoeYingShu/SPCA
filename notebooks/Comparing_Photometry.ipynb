```python
RMS = []
for i in range(len(Run_list)):
    # folder of the Run
    foldername = path1 + Run_list[i]

    # name of file where the data is 
    snip = 'ch1_datacube_binned_AORs464.dat'

    # get the data from .dat file
    flux, time, xdata, ydata, psfwx, psfwy = comp.get_binned_data(foldername, snip)

    # normalize flux
    flux_norm = flux/np.median(flux)

    # Smooth lightcurve
    smooth = comp.highpassflist(flux_norm)

    # get residuals
    residual = flux_norm - smooth 

    # calculate RMS
    RMS_tmp = comp.RMS(residual)
    RMS.append(RMS_tmp)

    # plotting
    path_plot = path1 + Run_list[i] + '/RMS_Scatter.pdf'

    fig, axes = plt.subplots(ncols = 1, nrows = 2, sharex = True, figsize = (10,6))
    fig.suptitle('RMS = '+ str(RMS_tmp))

    axes[0].plot(time, flux_norm, 'k.', alpha = 0.15, label='Measured Flux')
    axes[0].plot(time, smooth, '+', label = 'Filtered')
    axes[0].set_ylabel('Normalized Flux')
    axes[1].plot(time, residual, 'k.', alpha =0.1)
    axes[1].set_xlim(np.min(time), np.max(time))
    axes[1].axhline(y=0, linewidth = 1)
    axes[1].set_ylabel('Residuals')
    axes[1].set_xlabel('Time (BMJD)')
    fig.subplots_adjust(hspace=0)
    fig.savefig(path_plot)
    plt.close()

RMS_fcenter = RMS
```

```python
def RMS(x):
    n=len(x)
    x2 = np.multiply(x,x)
    sumx2 =np.sum(x2)
    x_RMS = np.sqrt((1.0/n)*sumx2)
    return x_RMS
```

```python
def get_RMS(Run_list):
    RMS_list = np.empty(len(Run_list))
    for i in range(len(Run_list)):
        foldername = Run_list[i]
        flux, flux_err, time, xdata, ydata, psfwx, psfwy = get_data(foldername)
        smooth = highpassflist(flux)
        residual = flux - smooth 
        RMS_list[i] = RMS(residual)
        path = Run_list[i] + '/RMS_Scatter.pdf'
        fig, axes = plt.subplots(ncols = 1, nrows = 2, sharex = True, figsize = (10,6))
        fig.suptitle('RMS = '+ str(RMS_list[i]))
        axes[0].plot(time, flux, 'k.', alpha = 0.15, label='Measured Flux')
        axes[0].plot(time, smooth, '+', label = 'Filtered')
        axes[0].set_ylabel('Raw Flux')
        axes[1].plot(time, flux-smooth, 'k.', alpha =0.1)
        axes[1].set_xlim(np.min(time), np.max(time))
        axes[1].axhline(y=0, color='b', linewidth = 1)
        axes[1].set_ylabel('Residual')
        axes[1].set_xlabel('Time since IRAC turn on(days)')
        fig.subplots_adjust(hspace=0)
        fig.savefig(path)
    return RMS_list
```
